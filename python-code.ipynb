```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
import statsmodels.api as sm
import statsmodels as statsmodels
```

```python
#Augmented Dickey-Fuller Test
#H0 = Data Non-Stasioner
#Ha = Data Stasioner
#Jika p-value < 0.05 maka Data Stasioner

from statsmodels.tsa.stattools import adfuller
result = adfuller(df['Average_Temp'])
print(result)
if result[1] < 0.05:
  print("Data Stasioner")
else:
  print("Data Non-Stasioner")
```

```python
#Plot ACF
plot_acf(df, lags = 50)
plt.ylim([-1,1])

#Plot PACF
plot_pacf(df, lags = 50)
plt.ylim([-1,1])
```

## Pemodelan



```python
#Split Data
jumlah_pred = 10
df_train = df[:-jumlah_pred].astype(float)
df_test = df[-jumlah_pred:].astype(float)
print(df_train)
print(df_test)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = sm.tsa.statespace.SARIMAX(df_train.astype(float),
                                  order=(1,0,2),
                                  seasonal_order=(1,0,1,6))
model_fit = model.fit()       #tahap estimasi parameter optimal

prediction = model_fit.forecast(steps=jumlah_pred, alpha=0.05)

#Memasukkan Nilai Prediksi ke Suatu Variabel
array_prediction = []
for i in range(len(prediction)):
  array_prediction.append(prediction.iloc[i])
  i = i + 1
```

```python
model_fit.summary()
```

```python
# Menampilkan parameter model yang digunakan
params = model_fit.params
print("Parameters of the SARIMAX model:")
print(params)
```

```python
array_prediction
```

```python
# RMSE_Percentage
array_test = np.array(df_test["Average_Temp"].astype(float))
RMSE_Percentage = np.abs(((array_test - array_prediction)/array_test)*100)
np.mean(RMSE_Percentage)
```

```python
# Mempersiapkan Grafik Test vs Prediction

df_prediction = df_test.copy()
df_prediction["Average_Temp"] = array_prediction
df_prediction["Average_Temp"] = df_prediction["Average_Temp"].astype(float)
df_prediction.info()
```

```python
# Plot Test vs Prediction
display_window = 100   # Time window for Test vs Prediction plot
plt.plot(np.array(df.index.values)[-display_window:], df['Average_Temp'][-display_window:], label='Test')
plt.plot(np.array(df_prediction.index.values), df_prediction['Average_Temp'], label='Prediction')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Test vs Prediction')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
